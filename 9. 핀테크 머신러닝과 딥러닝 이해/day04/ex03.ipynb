```python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
# 학습세트, 테스트세트 분리
import numpy as np
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target
)
```

```python
# 데이터 전처리 (표준 점수로 변환 - 학습 데이터가 한쪽으로 편향되는 것을 막기 위해서!)
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=10) # 로지스틱 함수 (이진 크로스 엔트로피 함수) OVR - One Versus Rest (한개와 나머지)
sc.fit(train_scaled, train_target) # 내부적으로 하나씩 꺼내서 사용 - 확률적 경사 하강법 

print(sc.score(train_scaled, train_target)) # 학습 세트
print(sc.score(test_scaled, test_target)) # 테스트 세트 
```

```python
# partial_fit(..) : 추가 학습할 때 사용하는 메서드 
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target)) # 학습 세트
print(sc.score(test_scaled, test_target)) # 테스트 세트
```

```python
sc = SGDClassifier(loss='log_loss', max_iter=10)

train_score  = []
test_score = []
```

```python
classes = np.unique(train_target)

for _ in range(0, 300): # _ : 자리만 차지하고 1회씩 반복 
    sc.partial_fit(train_scaled, train_target, classes=classes)

    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_score) # 훈련 세트 정확도
plt.plot(test_score) # 테스트 세트 정확도
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
# tol - 0.01 기본값, 0.01만큼 정확도가 향상안되면 종료
# tol=None : 향상되지 않으면 멈춘다 (=조기종료), 앞 횟수만큼은 무조건 반복함 
sc = SGDClassifier(loss='log_loss', max_iter=55, tol=None) 
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
import requests
url = "https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EB%B8%94%EB%A1%9C%EA%B7%B8"
response = requests.get(url)
html_text = response.text
```

```python
# 위 url에서 a태그에 있는 이미지 주소만 가져오기!
import re
pattern = r"<a.*href=['\"]([^'\"]+)['\"][^>]*>"
p = re.compile(pattern, re.I) # re.I == re.IGNORECASE
for m in p.finditer(html_text):
    print(m.group(1))
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(html_text, 'html.parser')
```

```python
# BeautifulSoup 사용하여 a태그에 있는 이미지 주소 가져오기 
items = soup.select("a.news_tit")
for item in items:
    href = item.attrs['href']
    print(href)
```

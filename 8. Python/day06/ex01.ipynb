```python
from faker import Faker
```

```python
faker = Faker() # 기본값은 영어 데이터
faker.name() # 랜덤값 출력
```

```python
faker = Faker("ko-KR") # 언어코드-국가코드
faker.name() # 랜덤값 출력
```

```python
# 천명의 회원데이터 출력하기 - 가짜 데이터 
members = [(faker.email(), faker.name(), faker.address()) 
               for _ in  range(0, 1000)] # for문에서 i 변수가 사용되지 않는 경우 언더바(_)로 자리만 채워줘도 사용 가능 
```

```python
for member in members:
    print(member)
```
